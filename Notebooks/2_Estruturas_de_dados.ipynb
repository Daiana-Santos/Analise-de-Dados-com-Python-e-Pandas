# **Tuplas**

```python
#as tuplas usam () como sintaxe
tp = ("Banana", "Maçã", 10, 50)
tp
```

```python
#retornando o primeiro elemento
tp [0]
```

```python
#diferente das listas as tuplas são imutáveis, o que quer dizer que podemos mudar seus elementos
tp[0]= "Laranja"
```

# **Dicionários**

```python

#Para criar um dicionário utilizamos a {}
dc = {"Maçã":20,"Banana":10, "Laranja":15, "Uva":5} #Dicionários trabalham com o conceito chave e valor
dc
```

```python
#acessando um dicionário através de chave
dc ["Maçã"]
```

```python
#atualizando o valor e maçã
dc ["Maçã"] = 25
dc
```

```python
#retornando todas as chaves do dicionário
dc.keys()
```

```python
#retornando valores do dicionário
dc.values()
```

```python
#verificando se ja exixte uma chave no dicionário e caso não exista inserir
dc.setdefault("Limão",22)
```
